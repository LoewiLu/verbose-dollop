```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data=np.arange(100,201) #从100-200的100个数
plt.plot(data)
plt.show()

x = np.arange(20)
y = x**2
plt.plot(x, y) #绘制方程式 y=x^2
plt.show()
```

```python
# 多个figure
data1=np.arange(100,201)
plt.plot(data1)

data2=np.arange(200.301)
plt.figure()#用此画多个图
plt.plot(data2)

data3=np.arange(600.701)
plt.figure()#用此画多个图
plt.plot(data3)

plt.show()
```

```python
#多个subplot
data4=np.arange(100,201)
plt.subplot(2,1,1)#一个窗口显示多个图形，2行1列中的第1个subplot
plt.plot(data4)

data5=np.arange(200.301)
plt.subplot(2,1,2)#一个窗口显示多个图形，2行1列中的第2个subplot
plt.plot(data5)

plt.show()
```

```python
#线性图
# plt.plot([横轴值],[纵轴值],'-r 红色直线'/':g'绿色点线)
plt.plot([10,20,30],[30,60,90],':r')
plt.plot([10,20,30],[20,30,90],'-g')

plt.show()
```

```python
#散点图
N=20 #20个
plt.scatter(np.random.rand(N)*100,
           np.random.rand(N)*100,
           c='r',s=100,alpha=0.5)
# c=color,s=size,alpha=transparent

plt.scatter(np.random.rand(N)*100,
           np.random.rand(N)*100,
           c='g',s=200,alpha=0.5)

plt.scatter(np.random.rand(N)*100,
           np.random.rand(N)*100,
           c='b',s=300,alpha=0.5)
plt.show()
```

```python
#饼状图
labels = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

data = np.random.rand(7)*100 #包含7个数据的随机数值

plt.pie(data,labels=labels,autopct="%1.2f%%")#autopct 数值的精准度 .1保留一位小数
plt.axis('equal') #坐标轴大小一致
plt.legend() #绘制图例 

plt.show()
```

```python
#条形图

N=7 #7个数

x = np.arange(N) 
data=np.random.randint(low=0,high=100,size=N) #每个数值是[0, 100]的随机数
colors=np.random.rand(N*3).reshape(N,-1)
#表示先生成21（N x 3）个随机数，然后将它们组装成7行，那么每行就是三个数，这对应了颜色的三个组成部分
labels=['Mon','Tue','Wed','Thu','Fri','Sat','Sun']

plt.title('Weekday Data')
plt.bar(x,data,alpha=0.6,color=colors, tick_label=labels)
plt.show()
```

```python
#直方图
data=[np.random.randint(0,n,n)for n in [3000,4000,5000]]
#包含了三个数组的数组
#第一个数组包含了3000个随机数，这些随机数的范围是 [0, 3000)
#第二个数组包含了4000个随机数，这些随机数的范围是 [0, 4000)
#第三个数组包含了5000个随机数，这些随机数的范围是 [0, 5000)

labels=['3K','4K','5K']

bins=[0,100,500,1000,2000,3000,4000,5000]
#bins直方图的边界：[0, 100) 会有一个数据点，[100, 500)会有一个数据点，。。。
#7个逗号就是说7个数据点

plt.hist(data,bins=bins,label=labels)
plt.legend()

plt.show()

#三组数据在3000以下都有数据，并且频度是差不多的
#但蓝色条只有3000以下的数据，橙色条只有4000以下的数据
#与我们的随机数组数据刚好吻合
```
